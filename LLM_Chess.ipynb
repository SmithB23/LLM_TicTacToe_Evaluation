```python
import re
import json
from transformers import pipeline, set_seed

# Initialize model
generator = pipeline("text-generation", model="distilgpt2")
set_seed(42)

def parse_prompt_grid(prompt):
    lines = prompt.strip().split("\n")
    board = []
    for i in range(1, 4):
        row = []
        for cell in lines[i].split('|'):
            c = cell.strip()
            row.append(c if c in ['X', 'O'] else '.')
        board.append(row)
    player_line = [line for line in lines if "Player" in line]
    player = 'X' if player_line and '1' in player_line[0] else 'O'
    return board, player

def board_to_coordinates(board):
    X_coords, O_coords, Empty_coords = [], [], []
    for r in range(3):
        for c in range(3):
            if board[r][c] == "X":
                X_coords.append(f"({r},{c})")
            elif board[r][c] == "O":
                O_coords.append(f"({r},{c})")
            else:
                Empty_coords.append(f"({r},{c})")
    return X_coords, O_coords, Empty_coords

def board_to_int(board):
    mapping = {'.':0, 'X':1, 'O':2}
    return [[mapping[cell] for cell in row] for row in board]

def test_won(board):
    size = 3
    def all_same(lst):
        return lst[0] != 0 and all(x == lst[0] for x in lst)
    # Rows, cols, diagonals
    for row in board:
        if all_same(row):
            return row[0]
    for col in zip(*board):
        if all_same(col):
            return col[0]
    if all_same([board[i][i] for i in range(size)]):
        return board[0][0]
    if all_same([board[i][size - 1 - i] for i in range(size)]):
        return board[0][size - 1]
    if any(0 in row for row in board):
        return -1
    return 0

def get_available_moves(board):
    return [(r,c) for r in range(3) for c in range(3) if board[r][c] == 0]

def minimax(board, player):
    winner = test_won(board)
    if winner != -1:
        if winner == player:
            return 1
        elif winner == 0:
            return 0
        else:
            return -1
    scores = []
    for move in get_available_moves(board):
        r,c = move
        board[r][c] = player
        score = -minimax(board, 3 - player)
        board[r][c] = 0
        scores.append(score)
    return max(scores) if scores else 0

def find_best_move(board, player):
    best_score = -float('inf')
    best_move = None
    for move in get_available_moves(board):
        r,c = move
        board[r][c] = player
        score = -minimax(board, 3 - player)
        board[r][c] = 0
        if score > best_score:
            best_score = score
            best_move = move
    return best_move

# Load dataset from file
def load_dataset(path):
    with open(path, "r") as f:
        return [json.loads(line) for line in f]

dataset = load_dataset("tictactoe_bestmove.jsonl")

for i, item in enumerate(dataset[:5]):
    prompt = item["prompt"]
    ground_truth = item["completion"]

    board, player_char = parse_prompt_grid(prompt)
    X_coords, O_coords, Empty_coords = board_to_coordinates(board)
    board_int = board_to_int(board)
    player_num = 1 if player_char == 'X' else 2

    # Prepare prompt with explicit empty spots list
    formatted_prompt = f"""X: {', '.join(X_coords) if X_coords else 'none'}
O: {', '.join(O_coords) if O_coords else 'none'}
Empty: {', '.join(Empty_coords) if Empty_coords else 'none'}
Player: {player_char}
Best move (row, col) is:
"""

    # Generate model prediction
    response = generator(
        formatted_prompt,
        max_new_tokens=6,
        do_sample=False,
        eos_token_id=generator.tokenizer.eos_token_id
    )[0]["generated_text"]

    # Extract predicted move
    match = re.search(r"\(\s*(\d)\s*,\s*(\d)\s*\)", response)
    predicted_move = (int(match.group(1)), int(match.group(2))) if match else None

    # Validate predicted move
    valid_moves = get_available_moves(board_int)
    valid = predicted_move in valid_moves if predicted_move else False

    if not valid:
        final_move = find_best_move(board_int, player_num)
        note = "[INVALID MOVE - replaced with minimax best move]"
    else:
        final_move = predicted_move
        note = ""

    # --- NEW: validate and fix ground truth ---
    gt_match = re.search(r"\(\s*(\d)\s*,\s*(\d)\s*\)", ground_truth)
    gt_move = (int(gt_match.group(1)), int(gt_match.group(2))) if gt_match else None
    if gt_move not in valid_moves:
        gt_move = find_best_move(board_int, player_num)  # fix to minimax
        gt_note = "[INVALID GT - replaced with minimax best move]"
    else:
        gt_note = ""

    print(f"\n=== Example {i+1} ===")
    print(f"X: {', '.join(X_coords) if X_coords else 'none'}")
    print(f"O: {', '.join(O_coords) if O_coords else 'none'}")
    print(f"Empty: {', '.join(Empty_coords) if Empty_coords else 'none'}")
    print(f"Player: {player_char}")
    print(f"Best move (row, col) is: {final_move} {note}")
    print(f"(Ground Truth: {gt_move}) {gt_note}")
```
